```python
import numpy as np

x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

# Line style and color

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100],
        [0, 0, 100, 100, 0, 100, 130, 100, 0])

plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100],
        [0, 0, 100, 100, 0, 100, 130, 100, 0],
        "g--")  # green dashed line

plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-",
         [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x, "g--")
plt.plot(x, x**2, "r:")
plt.plot(x, x**3, "b^")
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, "g--", x, x**2, "r:", x, x**3, "b^")
line1.set_linewidth(3.0)
line2.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

# Saving a figure

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x**2)
plt.savefig("my_square_function.png", transparent=True)
```

# Subplots

```python
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 cols, 1st subplot = top left
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 rows, 2 cols, 2nd subplot = top right
plt.plot(x, x**2)
plt.subplot(2, 2, 3)
plt.plot(x, x**3)
plt.subplot(2, 2, 4)
plt.plot(x, x**4)
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, x)
plt.subplot(2, 2, 2)
plt.plot(x, x**2)
plt.subplot(2, 1, 2)
plt.plot(x, x**3)
plt.show()
```

```python
plt.subplot2grid((3, 3), (0,0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3, 3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3, 3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3, 3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

# Multiple figures

```python
x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Square and Cube")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y  x**5")

plt.figure(1)
plt.plot(x, -x**3, "r:")

plt.show()
```
