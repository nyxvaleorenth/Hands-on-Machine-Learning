# Reshaping an array

## In place

```python
g = np.arange(24)
print(g)
print("Rank:", g.ndim)
```

```python
g.shape = (6,4)
print(g)
print("Rank:", g.ndim)
```

```python
g.shape = (2, 3, 4)
print(g)
print("Rank:", g.ndim)
```

## `reshape`
return new `ndarray` pointing at the same data. this means that modifying one array will change the other

```python
g2 = g.reshape(4, 6)
print(g2)
print("Rank:", g2.ndim)
```

```python
g2[1,2] = 999
g2
```

# Arithmetic operations

```python
a = np.array([14, 23, 32, 41])
b = np.array([5, 4, 3, 2])
print("a + b = ", a + b)
print("a - b = ", a - b)
print("a * b = ", a * b)  # not matrix multiplication
print("a / b = ", a / b)
print("a // b = ", a // b)
print("a % b = ", a % b)
print("a ** b = ", a ** b)
```

# Broadcasting

## First rule

```python
h = np.arange(5).reshape(1, 1, 5)
h
```

```python
h + [10, 20, 30, 40, 50]  # same as: h + [[[10, 20, 30, 40, 50]]]
```

## Second rule

```python
k = np.arange(6).reshape(2, 3)
k
```

```python
k + [[100], [200]]  # same as: k + [[100, 100, 100], [200, 200, 200]]
```

```python
# combine rule 1 & 2
k + [100, 200, 300]
# after rule 1: [[100, 200, 300]]
# after rule 2: [[100, 200, 300], [100, 200, 300]]
```

```python
k + 1000
# [[1000]]  k: 2,3  1000: 2,1
# [[1000, 1000, 1000], [1000, 1000, 1000]]
```

## Third rule

```python
# after rules 1 & 2, the sizes of all arrays must match
try:
    k + [33, 44]
except ValueError as e:
    print(e)
```

## Upcasting

```python
k1 = np.arange(0, 5, dtype=np.uint8)
print(k1.dtype, k1)
```

```python
k2 = k1 + np.array([5, 6, 7, 8, 9], dtype=np.int8)
print(k2.dtype, k2)
# upcasted to in16 to represent both int and uint 8
```

```python
k3 = k1 + 1.5
print(k3.dtype, k3)
# upcasted to float
```

# Mathematical and statistical functions

## `ndarray` methods

```python
a = np.array([[-2.5, 3.1, 7], [10, 11, 12]])
print(a)
print("mean =", a.mean())
```

```python
for func in (a.min, a.max, a.sum, a.prod, a.std, a.var):
    print(func.__name__, "=", func())
```

```python
c = np.arange(24).reshape(2,3,4)
c
```

```python
c.sum(axis=0)  # sum across matrices
```

```python
c.sum(axis=1)  # sum across rows
```

```python
c.sum(axis=(0,2))  # sum across matrices and columns
```

## Universal functions (ufunc)

```python
a = np.array([[-2.5, 3.1, 7], [10, 11, 12]])
np.square(a)
```

```python
print("Original ndarray")
print(a)
for func in (np.abs, np.sqrt, np.exp, np.log, np.sign, np.ceil, np.modf, np.isnan, np.cos):
    print("\n", func.__name__)
    print(func(a))
```

## Binary ufuncs

```python
a = np.array([1, -2, 3, 4])
b = np.array([2, 8, -1, 7])
np.add(a, b)  # a + b
```

```python
np.greater(a, b)  # a > b
```

```python
np.maximum(a, b)
```

```python
np.copysign(a, b)  # copy the sign of the second array
```
